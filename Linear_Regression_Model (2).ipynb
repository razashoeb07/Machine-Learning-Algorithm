## Importing necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("C:\\Users\\shoeb\\\Desktop\\\homeprices.csv")
```

```python
df.head()
```

```python
plt.scatter(df["area"],df["price"],color = "red",marker = "+")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
new_df = df.drop("price",axis = "columns")
new_df
```

```python
model = linear_model.LinearRegression()
model
```

```python
model.fit(new_df,df.price)
```

### (1) Predict price of a home with area = 3300 sqr ft 

```python
model.predict([[3300]]) # new input
```

### (2) Predict price of a home with area = 5000 sqr ft 

```python
model.predict([[5000]]) # new input
```

### (3) Predict price of a home with area = 3600 sqr ft

```python
model.predict([[3600]]) # new input
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# y = m*x + b

167.30954677*3600 + 76692.3818707813
```

### Showing the Regression Line (Best Fit Line) 

```python
plt.scatter(df["area"],df["price"])
plt.plot(new_df,model.predict(new_df),color = "red")
plt.xlabel("area")
plt.ylabel("price")
plt.show()
```
